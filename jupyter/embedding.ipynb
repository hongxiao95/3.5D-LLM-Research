```python
import os
from openai import AzureOpenAI
import numpy as np
```

```python
apikey = ""
endpoint = ""
with open("azure_ai_jpease.apikey") as api_key_file:
    endpoint = api_key_file.readline().strip()
    apikey = api_key_file.readline().strip()
api_version = "2023-05-15"
```

```python
client = AzureOpenAI(api_key=apikey, api_version=api_version, azure_endpoint=endpoint)
```

```python
sentence_input = ["Do you hear the people sing","Do you hear the people  sing","起来不愿做奴隶的人们","你可听见他们唱","你的饭狗都不吃","彼之歌声，汝岂不闻？","你听见人民唱了吗","你看见他们唱歌了吗"]

response1 = client.embeddings.create(input=sentence_input, model="embedding_3")
```

```python
origin_embedding = np.array(response1.data[0].embedding,dtype=np.float64)
other_embeddings = [np.array(x.embedding, dtype=np.float64) for x in response1.data[1:]]
print(list(origin_embedding))
```

```python
print("下面给出各句话相对于Do you hear the people sing的相似度")
for i in range(len(other_embeddings)):
    print(f"Similarity: {np.dot(origin_embedding, other_embeddings[i])} at {sentence_input[i + 1]}")
```
